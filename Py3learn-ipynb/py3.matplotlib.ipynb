## 1. 简单的图形绘制

```python
# 绘图的基本流程
import matplotlib.pyplot as plt # 导入库中的pyplot函数，并给出缩写
plt.plot([1,2,3,4,5],[5,6,5.5,7,8]) # 调用plot绘制散点图，参数为x值和y值的数组
plt.show()                      # 展示绘制结果
```

```python
# 图例、轴标签、图标题
import matplotlib.pyplot as plt
import matplotlib
# 设置中文字体
matplotlib.rcParams['axes.unicode_minus'] = False
matplotlib.rcParams['font.sans-serif'] = ['Microsoft YaHei']
matplotlib.rcParams['font.family']='sans-serif'
x = [1,2,3,4,5]         # 传入第一个曲线的x坐标和y值
y = [5,7,4,6,8]
x2 = [1,2,3,4,5]        # 传入第一个曲线的x坐标和y值
y2 = [10,14,12,10,13]
plt.plot(x, y, label='第一条线')    # 根据x、y对绘制曲线，加图例标签
plt.plot(x2, y2, label='第二条线')
plt.xlabel('x轴标签')               # 绘制坐标轴标签
plt.ylabel('y轴标签')
plt.title('折线图图标题')           # 绘制图标题
plt.legend()        # 展示图例
plt.show()          # 展示绘图结果
```

## 2. 条形图

```python
import matplotlib.pyplot as plt
import matplotlib
# 设置中文字体
matplotlib.rcParams['axes.unicode_minus'] = False
matplotlib.rcParams['font.sans-serif'] = ['Microsoft YaHei']
matplotlib.rcParams['font.family']='sans-serif'
# 绘制柱状图，注意要成组出现，所有x坐标就要衔接恰当
plt.bar([1,5,9,13,17],[2,3,5,4,7],label='条形图1')  # color默认蓝色
plt.bar([2,6,10,14,18],[1,4,6,3,8],label='条形图1',color='y')
plt.bar([3,7,11,15,19],[3,6,8,12,5],label='条形图3',color='r')
plt.xlabel('数量')                  # 绘制坐标轴标签
plt.ylabel('高度')
plt.title('条形图图标题')            # 绘制图标题
plt.legend()        # 展示图例
plt.show()          # 展示绘图结果
```

## 3. 直方图

```python
import matplotlib.pyplot as plt
import matplotlib
# 设置中文字体
matplotlib.rcParams['axes.unicode_minus'] = False
matplotlib.rcParams['font.sans-serif'] = ['Microsoft YaHei']
matplotlib.rcParams['font.family']='sans-serif'
population_ages = [22,55,62,45,21,22,34,42,42,4,99,102,110,120,121,122,130,111,115,112,80,75,65,54,44,43,42,48]         # 年龄数据
bins = [0,20,40,60,80,100,120,140]      # 20岁为一个年龄区间
plt.hist(population_ages, bins, histtype='bar', rwidth=0.8)
plt.xlabel('x轴标签')         # 绘制坐标轴标签
plt.ylabel('y轴标签')
plt.title('直方图图标题')# 绘制图标题
plt.legend()            # 展示图例
plt.show()              # 展示绘图结果
```

## 4. 散点图

```python
import matplotlib.pyplot as plt
import matplotlib
# 设置中文字体
matplotlib.rcParams['axes.unicode_minus'] = False
matplotlib.rcParams['font.sans-serif'] = ['Microsoft YaHei']
matplotlib.rcParams['font.family']='sans-serif'
x = [1,2,3,4,5,6,7,8]   # x坐标数组
y = [5,2,4,2,1,4,5,2]   # 对应的y值数组
plt.scatter(x,y, label='散点', color='g', s=90, marker=".") # 绘制scatter散点图
# label图例名、color颜色、s直径、marker标记图形
plt.xlabel('x轴标签')         # 绘制坐标轴标签
plt.ylabel('y轴标签')
plt.title('散点图图标题')# 绘制图标题
plt.legend()            # 展示图例
plt.show()              # 展示绘图结果
```

## 5. 堆叠图

```python
import matplotlib.pyplot as plt
import matplotlib
# 设置中文字体
matplotlib.rcParams['axes.unicode_minus'] = False
matplotlib.rcParams['font.sans-serif'] = ['Microsoft YaHei']
matplotlib.rcParams['font.family']='sans-serif'
days = [1,2,3,4,5]      # 天数相当于x
sleeping = [7,8,6,11,7] # 分别录入4类数据的数组
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]
plt.plot([],[],color='m', label='睡觉', linewidth=5) # 预设4类数据的属性
plt.plot([],[],color='c', label='吃饭', linewidth=5)
plt.plot([],[],color='r', label='工作', linewidth=5)
plt.plot([],[],color='g', label='玩耍', linewidth=5)
plt.stackplot(days, sleeping, eating, working, playing, colors=['m','c','r','g']) # 绘制堆积图，依次列出一个自变量和所有因变量
plt.xlabel('x轴标签')         # 绘制坐标轴标签
plt.ylabel('y轴标签')
plt.title('堆积图图标题')# 绘制图标题
plt.legend()            # 展示图例
plt.show()              # 展示绘图结果
```

## 6. 饼图

```python
import matplotlib.pyplot as plt
import matplotlib
# 设置中文字体
matplotlib.rcParams['axes.unicode_minus'] = False
matplotlib.rcParams['font.sans-serif'] = ['Microsoft YaHei']
matplotlib.rcParams['font.family']='sans-serif'
slices = [7,2,2,13]         # 扇区数值
activities = ['睡觉','吃饭','工作','玩耍']  # 扇区标签
cols = ['c','m','r','b']    # 扇区颜色
plt.pie(slices,             # 绘制，数据=slices，标签=activities，颜色=colors
    labels=activities,
    colors=cols,
    startangle=90,          # 起始绘制角度90度
    shadow= True,           # 图形带阴影
    explode=(0, 0.1, 0, 0), # 第二项突出显示（吃饭）
    autopct='%1.1f%%')      # ？？？
plt.title('饼图图标题')      # 图标题
plt.show()                  # 展示绘图结果
```

## 7. 从文件加载数据

```python
import matplotlib.pyplot as plt
import matplotlib
import csv
# 设置中文字体
matplotlib.rcParams['axes.unicode_minus'] = False
matplotlib.rcParams['font.sans-serif'] = ['Microsoft YaHei']
matplotlib.rcParams['font.family']='sans-serif'
x = []      # x数据入口
y = []      # y数据入口
with open('example.txt','r') as csvfile:        # 以只读方式打开文本文件，读入后命名为csvfile
    plots = csv.reader(csvfile, delimiter=',')  # 以“，”为分隔符，用csv.reader读入数据到plots变量
    for row in plots:
        x.append(int(row[0]))                   # 从plots变量，逐步提取x和y的值
        y.append(int(row[1]))
plt.plot(x,y, label='从文件加载数据!')          # 绘制图形
plt.xlabel('x轴标签')   # 绘制坐标轴标签
plt.ylabel('y轴标签')
plt.title('图标题')     # 绘制图标题
plt.legend()            # 展示图例
plt.show()              # 展示绘图结果
```

## 8. 使用numpy从文件加载数据

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
# 设置中文字体
matplotlib.rcParams['axes.unicode_minus'] = False
matplotlib.rcParams['font.sans-serif'] = ['Microsoft YaHei']
matplotlib.rcParams['font.family']='sans-serif'
x, y = np.loadtxt('example.txt', delimiter=',', unpack=True) # 导入数据
plt.plot(x,y, label='从numpy文件加载数据!')          # 绘制图形
plt.xlabel('x轴标签')   # 绘制坐标轴标签
plt.ylabel('y轴标签')
plt.title('图标题')     # 绘制图标题
plt.legend()            # 展示图例
plt.show()              # 展示绘图结果
```
